```python
import re

def load_file(text):
    # Match headings of the form [heading] and capture everything under them
    pattern = r"\[([^\]]+)\]\s*((?:.|\n)*?)(?=\n\[|$)"
    
    matches = re.findall(pattern, text)
    result = {}
    for heading, content in matches:
        result[heading.strip()] = content.strip()
    return result
```
